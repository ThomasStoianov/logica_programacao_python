Desafio: Processamento de Dados de Vendas
Você recebeu uma carga de dados bruta de um sistema de vendas.
Os dados estão estruturados como **lista de tuplas**, onde cada tupla
representa um registro:
(ID_Produto, Nome, Quantidade, Preço_Unitário).
**Seu objetivo:**
1. Filtrar produtos com estoque baixo (Quantidade < 10);
2. Calcular o valor total do inventário;
3. Gerar uma lista de tuplas atualizada com um reajuste de preço.
**Dados Iniciais:**
vendas_brutas = [(101, "Monitor", 5, 1200.0), (102, "Mouse", 50, 80.0), (103,
"Teclado", 8, 150.0), (104, "Case", 3, 50.0)]
